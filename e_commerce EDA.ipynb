```python

df['Returns'].fillna(0,inplace =True)
```

```python
def return_vlue (values):
    values=str(values).split('.')[0]
    return int(values)
df['Returns'] = df['Returns'].apply(return_vlue)
```

```python
# convert Purchase date to_datetime 

df['Purchase Date']=pd.to_datetime(df['Purchase Date'])


df['year']=df['Purchase Date'].dt.year

df['month']=df['Purchase Date'].dt.month

df['week_day']=df['Purchase Date'].dt.day_name()

df['hour']=df['Purchase Date'].dt.hour
```
